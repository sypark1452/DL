## Graph Visualization
### Load Packages

```python
import numpy as np
import matplotlib.pyplot as plt 

# magic function
# jupyter 내부에 띄움
%matplotlib inline
```

### basic attributes

* alpha : 투명도
* kind : 그래프 종류, 'line', 'bar', 'barth', 'kde'
* logy : y축에 대해 log scailing
* use_index : 객체의 색인을 눈금 이름으로 사용할지 여부
* rot : 눈금 이름 돌리기 rotating 0 - 360
* xticks, yticks : x,y축으로 사용할 값
* xlim, ylim : x,y축의 한계
* gird : 축의 그리드를 표현할지 여부
* subplot : 각 column에 독립된 subplot 그리기
* sharex, sharey : subplot = True 이면 같은 x,y축을 공유하고 눈금과 한계를 연결
* figsize : 생성될 그래프의 크기를 tuple로 저장
* title : 그래프의 제목 저장
* legend : subplot의 범례 지정
* sort_columns : sort_columns를 알파벳 순서로 그린다

### matplotlib 사용하기

```python
# 점선 그래프 그리기
data = np.random.randn(50).cumsum()
data
```

```python
plt.plot(data)
plt.show()
```

```python
# 여러 그래프 그릴 준비 하기
plt.subplot(1,2,1)
plt.subplot(1,2,2)
plt.show()
```

```python
# multi 그래프 그리기
hist_data = np.random.randn(100)
scat_data = np.arange(30)
```

```python
plt.subplot(2,2,1)
plt.plot(data)
plt.subplot(2,2,2)
plt.hist(hist_data, bins = 20)
plt.subplot(2,2,3)
plt.scatter(scat_data, np.arange(30) + 3* np.random.randn(30) )
plt.show()
```

```python
# 그래프 선 옵션
# 색상 : b(파란색), g(초록색), r(빨간색), c(청록색), y(노란색), k(검은색), w(흰색)
# 마커 : o(원), v(역삼각형), ^(삼각형), s(네모), +(플러스), .(점)
```

```python
plt.plot(data, 'y^') # 색상 - 마커
plt.show()
```

```python
# 그래프 사이즈 조절
plt.figure(figsize=(5,5)) # inch 단위
plt.plot(data, 'k*')
plt.show()
```

```python
# 그래프 겹치기 + legend 달기
data = np.random.randn(30).cumsum()

plt.plot(data, 'k--', label = 'Default')
plt.plot(data, 'k-', drawstyle = 'steps-post', label = 'steps-post')
plt.legend()
plt.show()
```

```python
# 이름 달기
plt.plot(np.random.randn(1000).cumsum())
plt.title('Random Graph')
plt.xlabel('Stages')
plt.ylabel('Values')
plt.show()
```

```python
# 종합
plt.plot(np.random.randn(1000).cumsum(), 'k', label = 'one')
plt.plot(np.random.randn(1000).cumsum(), 'b', label = 'two')
plt.legend()
plt.show()
```

```python
# 그래프 저장하기
plt.savefig('시각화기초.svg')
```

## 이미지 시각화
### package 불러오기

```python
import numpy as np 
from PIL import Image
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
# 이미지 파일 열기
path =  'cat.jpg'

image_pil = Image.open(path)
image = np.array(image_pil)
```

```python
image.shape
```

```python
#   이미지 들여다 보기
np.min(image), np.max(image)
```

```python
# 그래프로 시각화하기
plt.hist(image.ravel(), 250, [0, 250])
plt.show()
```

```python
# 그림 나타내기
plt.imshow(image)
plt.show()
```

```python
# 이미지 흑백으로 열기
image_pil = Image.open(path).convert("L")
image_bw = np.array(image_pil)
image_bw.shape
```

```python
# 흑백 이미지 열기
plt.imshow(image_bw, 'gray')
plt.show()
```

```python
# 다른 색상으로 cmap 표현하기
# gray scale
```

```python
# RdBu
plt.imshow(image_bw, 'RdBu')
plt.show()
```

```python
# jet
plt.imshow(image_bw, 'jet')
plt.show()
```

```python
# colorbar 추가하기
plt.imshow(image_bw, 'jet')
plt.colorbar()
plt.show()
```

```python
# 이미지 설정
# 이미지 보기 사이즈 조절
plt.figure(figsize=(7,7))
plt.imshow(image)
plt.show()
```

```python
# 이미지에 제목 추가
plt.title('Cat')
plt.imshow(image)
plt.show()
```
